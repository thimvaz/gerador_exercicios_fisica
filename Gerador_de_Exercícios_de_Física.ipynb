```python
# Exercício de mistura (3 amostras) em ambiente adaiabático sem mudança de fase

import random

contador = 7
total = 10
# criar um laço para gerar n problemas com resultados inteiros
while contador <= total :
  # Gerando valores
  c_A = float(random.randrange(1, 11, 1)) / 10
  c_B = float(random.randrange(1, 11, 1)) / 10
  c_C = int((20 - c_A * 10 - c_B * 10))/10

  m_A = random.randrange(50, 1050, 25)
  m_B = random.randrange(50, 1050, 25)
  m_C = 2000 - m_A - m_B

  ti_A = random.randrange(0, 120, 5)
  ti_B = random.randrange(0, 120, 5)
  ti_C = 200 - ti_A - ti_B
      

  # fazer uma condição com m_C e c_C maiores que zero para criar exercício
  if (m_C > 0 and c_C > 0 and m_A != m_B != m_C and ti_A != ti_B != ti_C and c_A != c_B != c_C):
    
    t_eq = (m_A * c_A * ti_A + m_B * c_B * ti_B + m_C * c_C * ti_C) / (m_A * c_A + m_B * c_B + m_C * c_C)
    # fazer uma condição com  t_eq % 1 == 0.0 para criar exercício
    if (t_eq % 1 == 0.0):
      print('{}) Três amostras A, B e C são colocadas em contato dentro de um recipiente adiabático.'.format(contador))
      print('A amostra A apresenta massa de {}g, calor específico de {}cal/(g°C) e temperatura inicial de {}°C.'.format(m_A, c_A, ti_A))
      
      print('A amostra B apresenta massa de {}g, calor específico de {}cal/(g°C) e temperatura inicial de {}°C.'.format(m_B, c_B, ti_B))
      
      print('A amostra C apresenta massa de {}g, calor específico de {}cal/(g°C) e temperatura inicial de {}°C.'.format(m_C, c_C, ti_C))
      print('')
      print('Resposta {}: {}°C'.format(contador, t_eq) )
      print('')
      print('')
      print('')
      contador = contador + 1
```

##Escala Termométrica Arbitrária


```python
from math import gcd
# Definindo um símbolo para a escala arbitrária
simbolo = input("Entre com um símbolo para a sua escala termométrica personalizada: ") 
print("")
print("Beleza, chamaremos sua escala de °{}".format(simbolo))
print("")
# Definindo Ponto de Fusão e Ponto de Ebulição
print("Agora vamos definir o ponto de fusão (PF) e ebulição (PE) da água na sua escala.")
print("Lembrete: O PE SEMPRE É MAIOR que o PF!")
print("")

pf = int(input("Entre com o valor do PF da água na sua escala: ")) 
pe = int(input("Entre com o valor do PE da água na sua escala: "))
print("")
print("Beleza. Na sua escala, o PF = {}°{} e o PE = {}°{}".format(pf, simbolo, pe, simbolo))
print("")



```

```python
# Obtendo as equações de Conversão de Escalas
div_escala = pe-pf
# °C (PF = 0°C    PE = 100° C   PF-PE = 100)
div_geral_C = gcd(100, div_escala) 
print("A escala de conversão entre °{} e °C(Célsius) é a seguinte:".format(simbolo))
print("")
print("[T(°{})-{}]/{} = T(°C)/{}".format(simbolo, pf, div_escala / div_geral_C, 100/div_geral_C ))
print("_____________________________________")

# °F (PF = 32°C    PE = 212° C   PF-PE = 180)
div_geral_F = gcd(180, div_escala)
print("A escala de conversão entre °{} e °F(Fahreinheit) é a seguinte:".format(simbolo))
print("")
print("[T(°{})-{}]/{} = [T(°F)-32]/{}".format(simbolo, pf, div_escala / div_geral_F, 180/div_geral_F ))
print("_____________________________________")

# °K (PF = 32°C    PE = 212° C   PF-PE = 180)
div_geral_K = gcd(100, div_escala)
print("A escala de conversão entre °{} e K(Kelvin) é a seguinte:".format(simbolo))
print("")
print("[T(°{})-{}]/{} = [T(K)-273]/{}".format(simbolo, pf, div_escala / div_geral_K, 100/div_geral_K ))
```
